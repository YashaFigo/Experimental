```python
import csv
import re
from wordcloud import WordCloud
```

```python
data = open('MobyDick.txt')
csvreader = csv.reader(data,delimiter="|")
moby = list(csvreader)

for x in moby:
    if len(x) == 0:
        moby.remove(x)

def clean(x):
    x = "".join(x)
    substitutes = ["\n|-|;|\?|\|\.|,|!|\"|\.|\'|\(|\)"]
    for z in substitutes:
        y = re.sub(z," ",x)
    return y.lower().split()
        
moby = [clean(x) for x in moby]

all_words = []

for x in moby:
    for y in x:
        all_words.append(y)

print(all_words[:50])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

wordcloud = WordCloud(max_font_size=90,
                      max_words = 200,
                      scale = 10).generate(
                      " ".join(all_words))

plt.figure(figsize = (30,30))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
