### LETS WRAP IT IN A FUNCTION

```python
def nth_biggest_item(matrix,nth_number):
    nth_list = []
    for col_name in matrix:
        column = [col_name for x in range(len(matrix))]
        index = [str(x) for x in matrix[col_name].index]
        data = [x for x in matrix[col_name]]
        data_zip = zip(column,index,data)
        sorter = sorted(data_zip,key=lambda x: x[2],reverse=True)
        nth_list.append(sorter[nth_number])
    return nth_list
nth_biggest_item(matrix,1)    
```

```python
largest_corr = sorted(nth_biggest_item(matrix,1),key=lambda x: x[2],reverse=True)[0]
largest_corr
```
