## PREPARING AND CLEANING THE DATA:

```python
%ls
```

```python
melania = [x.lower().split() for x in open('melania.txt')][0]
michelle = [x.lower().split() for x in open('michelle.txt')][0]
```

```python
print(melania)
```

```python
print(michelle)
```

```python
michelle = [x for x in michelle if x != '\xe2\x80\x94']
```

```python
print(michelle)
```

```python
import string
all_char = []

for x in michelle:
    for y in x.split():
        for z in y:
            all_char.append(z)

for x in melania:
    for y in x.split():
        for z in y:
            all_char.append(z)

all_char = set(all_char)
symbols = [x for x in all_char if x not in string.ascii_lowercase]
symbols
```

```python
michelle = ["".join(y for y in x if y not in symbols) for x in michelle]
melania = ["".join(y for y in x if y not in symbols) for x in melania]
```

```python
print(michelle)
```

```python
print(melania)
```

#### LETS COMPARE SETS:

```python
melania_set = set(melania)
michelle_set = set(michelle)
```

```python
counter = 0
for x in melania_set:
    if x in michelle_set:
        counter += 1
print("# FULL WORDS MELANIA STOLE",counter)
print("% STOLEN WORDS OUT OF ALL UNIQUE WORDS",((counter/len(melania_set))*100))
```
