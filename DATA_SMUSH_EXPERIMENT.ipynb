## SMUUUSHIN' DATA!

```python
data = pd.read_csv('admissions.csv')
data = data.drop('Unnamed: 0',axis=1).copy()
data[:10]
```

```python
plt.scatter(data['gpa'],data['gre'])
```

```python
from scipy.stats import linregress
slope, intercept, r_value, p_value, stderr_slope = linregress(data['gpa'],data['gre'])
```

```python
x_var = [x for x in data['gpa']]
y_var_linregress = [((slope*x)+intercept)for x in x_var]
```

```python
plt.scatter(data['gpa'],data['gre'])
plt.plot(x_var,y_var_linregress,color='r')
```

#### NOW LETS SMUSH!

```python
def smush_data(smush_factor):
    y_var_difference = []
    for x in range(len(y_var_linregress)):
        differential = data['gre'][x] - y_var_linregress[x]
        y_var_difference.append(differential)
    
    new_y_var = []
    for x in range(len(y_var_difference)):
        new_difference = y_var_difference[x]/smush_factor
        new_y_var.append(y_var_linregress[x] + new_difference)

    plt.scatter(data['gpa'],data['gre'])
    plt.scatter(data['gpa'],new_y_var,color='g')
    plt.plot(x_var,y_var_linregress,color='r')
    plt.show()  
```

----

#### Thee green data is a geometrically smushed version of the blue data about the regression line by the designated smush factor!

```python
smush_data(1)
```

```python
smush_data(1.2)
```

```python
smush_data(1.8)
```
